<b> Let us see the State Wise Analysis </b>

```python
plt.figure(figsize=(10,12))
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.xlabel('Population', fontsize=14)
plt.ylabel('States In India', fontsize=14)
plt.title("Population wrt to States in India", fontsize=16)
sns.barplot(x='TOTPOPULAT', y='STATNAME', data=df)  # Use x and y parameters to specify the data columns
plt.show()
```

```python
df.columns
```

```python
cols=['STATNAME','OVERALL_LI','FEMALE_LIT','MALE_LIT','SEXRATIO','DISTRICTS','TOTPOPULAT','P_URB_POP','P_RUR_POP','AREA_SQKM','SCHTOT','SCHTOTG','SCHTOTGR','SCHTOTPR','ENRTOT','ENRTOTG','ENRTOTGR', 
       'ENRTOTPR', 'TCHTOTG', 'TCHTOTP', 'SCLSTOT', 'STCHTOT', 'ROADTOT', 'SPLAYTOT', 'SWATTOT',  'SELETOT']
```

```python
df1=pd.DataFrame(df,columns=cols)
df1.head()
```

<b> Let us see how many schools are there State Wise </b>

```python
plt.figure(figsize=(10, 12))
sns.barplot(x='SCHTOT', y='STATNAME', data=df1)
plt.xlabel('No. of Schools', fontsize=14)
plt.ylabel('States in India', fontsize=14)
plt.title("Statewise schools in India", fontsize=16)
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.show()
```

<b> Let us see how many private and government schools are there in India </b>

```python
plt.figure(figsize=(8, 12))
sns.barplot(x='SCHTOTG', y='STATNAME', data=df1, alpha=0.8)
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.xlabel('No of GOVT Schools', fontsize=14)
plt.ylabel('States in India', fontsize=14)
plt.title("No of GOVT Schools wrt states in India", fontsize=16)
plt.show()
```

```python
plt.figure(figsize=(8, 12))
sns.barplot(x='SCHTOTPR', y='STATNAME', data=df1, alpha=0.8)
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.xlabel('No of private Schools', fontsize=14)
plt.ylabel('States in India', fontsize=14)
plt.title("No of private Schools wrt states in India", fontsize=16)
plt.show()
```

<b> Let us see the literacy rate wrt states </b>

```python
plt.figure(figsize=(8, 10))
sns.barplot(x='OVERALL_LI', y='STATNAME', data=df1, alpha=0.8)
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.xlabel('Literacy rate', fontsize=14)
plt.ylabel('States in India', fontsize=14)
plt.title("Literacy Rate wrt states in India", fontsize=16)
plt.show()
```

```python
plt.figure(figsize=(8, 10))
sns.barplot(x='FEMALE_LIT', y='STATNAME', data=df1, alpha=0.8)
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability
plt.xlabel('Female Literacy rate', fontsize=14)
plt.ylabel('States in India', fontsize=14)
plt.title("Female Literacy Rate wrt states in India", fontsize=16)
plt.show()
```

```python
plt.figure(figsize=(8,10))
sns.barplot(x='MALE_LIT', y ='STATNAME', data=df1, alpha=0.8)
plt.xticks(rotation='vertical')
plt.xlabel('Male Literacy rate', fontsize=14)
plt.ylabel('States in India', fontsize=14)
plt.title("Literacy Rate wrt states in India", fontsize=16)
plt.show()
```

<b> Let us see the top 5 and bottom 5 states which have the highest women literacy rate </b>

```python
top_5=df1.sort_values(by='OVERALL_LI', ascending=False).head(5)
bottom_5=df1.sort_values(by='OVERALL_LI', ascending=True).head(5)


top_bottom=pd.concat([top_5, bottom_5], axis=0, sort=False)
top_bottom
```

```python
plt.figure(figsize=(12, 8))
sns.barplot(x='OVERALL_LI', y='STATNAME', data=top_bottom, alpha=0.8)
plt.xticks(rotation=90)
plt.xlabel('Female Literacy Rate', fontsize=14)
plt.ylabel('States in India', fontsize=14)
plt.title("Comparison of Female Literacy Rate in Top 5 and Bottom 5 States", fontsize=16)
plt.show()
```

<b> Now let us see the which state has the highest difference in Male and Female Literacy </b>

```python
top_bottom['Difference']=top_bottom['MALE_LIT']-top_bottom['FEMALE_LIT']
top_bottom['Difference']
```

```python
plt.figure(figsize=(4,6))
sns.barplot(x='Difference', y='STATNAME', data=top_bottom, alpha=0.8)
plt.xticks(rotation='vertical')
plt.xlabel('Literacy rate difference', fontsize=14)
plt.ylabel('States in India', fontsize=14)
plt.title("the top  3 and bottom 3 states in difference literacy rate.", fontsize=16)
plt.show()
```

<b> Let us find out the RURAL Population in India </b>

```python
top_bottom['Rural']=100-top_bottom['P_URB_POP']
top_bottom['Rural']
```

```python
plt.figure(figsize=(10,8))

top_bottom.plot(y = ['P_URB_POP', 'Rural'], kind = 'bar')
plt.title("Comparison of Rural and Urban population")
```

<b> Now let us see the sex ratio </b>

```python
plt.figure(figsize=(12,8))
sns.barplot(x='STATNAME', y='SEXRATIO', data=top_bottom)
```
